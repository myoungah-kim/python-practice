## Cross Validation

총 데이터 개수가 적은 데이터셋의 경우 Train/Validation/Test 셋을 나누면 underfitting 문제가 생길 수 있음. 이럴 경우 5개, 10개 등으로 셋을 쪼개서 1 vs. REST로 학습&검증을 여러번 반복하는 법. 이렇게 여러번 반복해서 학습하는 방식으로 underfitting을 해결함.


1. K-Fold Cross Validation :  먼저 K 개의 데이터 폴드 세트를 만들어서 K번 만큼 각 폴드 세트에 학습과 검증 평가를 반복적으로 수행해서 평균 정확도로 모델을 평가.

2. Stratified K-Fold : When class in the data is highly imbalanced, it's a way of splitting the datasets while maintaining the ratio between classes. Even sampling.

3. cross_val_score() : Fold 세트 추출, 학습/예측, 평가 를 한꺼번에 수행.


### 1. K-Fold Cross Validation

1. K 결정
2. for 루프에서 반복적으로 학습/검증 데이터 추출 및 학습과 예측 수행
3. 폴드 세트 별로 예측 성능을 평균하여 최종 성능 평가

```
import warnings 
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

# Load data
path = '/content/drive/MyDrive/Colab Notebooks/DataPreprocess.csv'
df1 = pd.read_csv(path)

x = df1.values[:, :-1] # Country, Age, Salary (IV)
y = df1.values[:, -1] # Purchased (DV)
x, y
```

```
from sklearn.model_selection import KFold

cv = KFold(n_splits = 5, shuffle = True, random_state = 0)

for train_index, test_index in cv.split(x):
  print('train_index : ', train_index)
  print('.' * 80)
  print('test_index : ', test_index)
  print('=' * 80)
```

Iris Dataset 5 Fold Cross Validation

```
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import numpy as np

iris = load_iris()
features = iris.data
label = iris.target

dt_clf = DecisionTreeClassifier(random_state = 156)

kfold = KFold(n_splits = 5)
cv_accuracy = []
print('붓꽃 데이터 세트 크기 : ', features.shape[0])
```

```
# 5 fold cross-validation 할때마다 학습과 검증을 반복해 예측 정확도 측정

n_iter = 0

for train_index, test_index in kfold.split(features):

  # train & test split
  x_train, x_test = features[train_index], features[test_index]
  y_train, y_test = label[train_index], label[test_index]

  # train & predict
  dt_clf.fit(x_train, y_train)
  pred = dt_clf.predict(x_test)
  n_iter += 1

  # Calculate accuracy for every train-test repetition
  accuracy = np.round(accuracy_score(y_test, pred), 4)
  
  train_size = x_train.shape[0]
  test_size = x_test.shape[0]
  print('\n#{0} 교차 검증 정확도 : {1}, 학습 데이터 크기: {2}, 검증 데이터 크기: {3}'.format(n_iter, accuracy, train_size, test_size))
  print('#{0} 검증 세트 인덱스 : {1}'.format(n_iter, test_index))

  cv_accuracy.append(accuracy)

# Average accuracy
print('\n## 평균 검증 정확도 : ', np.mean(cv_accuracy))



```

## Feature Scaling & Standardize & Normalize

1. 표준화 Standardization (*StandardScaler*)

  z = (xi - mean)/std

  평균이 0 분산이 1인 Gaussian distribution 을 가진 값으로 변환. SVM, Linear Regression, Logistic Regression의 가정은 데이터가 정규분포를 이루고 있다는 하에 이루어지기 때문에 데이터를 표준화하는 것이 중요.


2. 정규화 Normalization (*MinMaxScaler*)

  (xi - min) / (max - min)

  데이터의 분포가 정규 분포가 아닐 경우 적용. 예를 들어 거리 0 ~ 100km & 0 ~ 100,000,000원 일때 최소 0~1로 변환.

ROC

http://scikit-learn.org/stable/auto_examples/model_selection/plot_roc.html

* ROC 곡선 자체는 FPR과 TPR의 변화값을 보는 데 이용
* AUC (Area Under Curve) 값: ROC 곡선 면적, 분류의 성능 지표, 1에 가까울수록 좋은 값.

```
# ROC curve

print(__doc__)

import numpy as np
import matplotlib.pyplot as plt
from itertools import cycle

from sklearn import svm, datasets
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
from scipy import interp

iris = datasets.load_iris()
x = iris.data
y = iris.target

# Binarize the output
y = label_binarize(y, classes = [0, 1, 2])
n_classes = y.shape[1]


# Add noisy features to make the problem harder
random_state = np.random.RandomState(0)
n_samples, n_features = x.shape
x = np.c_[x, random_state.randn(n_samples, 200 * n_features)]

# Shuffle and split training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .5, random_state = 0)

# Learn to predict each class against the other
clf = OneVsRestClassifier(svm.SVC(kernel = 'linear', probability = True, random_state = random_state))
y_score = clf.fit(x_train, y_train).decision_function(x_test)

# Compute ROC and AUC for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
  fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])
  roc_auc[i] = auc(fpr[i], tpr[i])

# Compute micro-average ROC curve and ROC area
fpr['micro'], tpr['micro'], _ = roc_curve(y_test.ravel(), y_score.ravel())
roc_auc['micro'] = auc(fpr['micro'], tpr['micro'])
```

```
plt.figure()
lw = 2
plt.plot(fpr[2], tpr[2], color = 'darkorange', lw = lw, label = 'ROC curve (area = %0.2f)' %roc_auc[2])
plt.plot([0, 1], [0, 1], color = 'navy', lw = lw, linestyle = '--')
plt.xlim(0.0, 1.0)
plt.xlim(0.0, 1.05)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Reciever operating characteristic example')
plt.legend(loc = 'lower right')
plt.show()
```

3 steps:

* Step 1. 각 데이터 포인트 i 에 대해 가장 가까운 중심점을 찾고, 그 중심점에 해당하는 클러스터를 할당. based on Euclidean distance.

* Step 2. 할당된 클러스터를 기반으로 새로운 중심점을 계산. 중심점은 클러스터 내부 점 좌표들의 mean.

* Step 3. 각 클러스터의 할당이 바뀌지 않을 때까지 반복.

## Evaluation (Silhouette Analysis)

Evaluation (Silhouette Analysis)

* 실루엣 값은 한 클러스터 안의 데이터들이 다른 클러스터와 비교해서 얼마나 비슷한가를 나타냄. 
* 같은 클러스터 내의 점들 간 거리는 가깝고, 서로 다른 클러스터 의 거리는 멀수록 높은 값을 얻을 수 있다.
* 일반적으로 실루엣 값이 0.5 보다 크다면 데이터가 잘 클러스터링 되었다는 것을 나타냄.

```
# Silhouette 
# 몇 개의 클러스터가 가장 좋은 실루엣 점수를 내는지 보기

from sklearn.metrics import silhouette_score

best_n = 1
best_score = -1

for n_cluster in range(2, 11):
  kmeans = KMeans(n_clusters = n_cluster)
  kmeans.fit(data)
  cluster = kmeans.predict(data)
  score = silhouette_score(data, cluster) ### 실루엣 점수

  print('클러스터의 수 : {}, 실루엣 점수 : {: .2f}'.format(n_cluster, score))

  if score > best_score :
    best_n = n_cluster
    best_score = score

print('가장 높은 실루엣 점수를 가진 클러스터 수 : {}, 실루엣 점수 : {: .2f}'.format(best_n, best_score))
```

# How to solve Overfitting / Underfitting

* Overfitting: High training score / Low test score
* Underfitting: Low training score / Low test score

How to solve Overfitting
1. 훈련 데이터를 더 많이 모음
2. Regularization : 규제, drop-out 등 다양한 방법을 이용해서 적당한 복잡도를 가지는 모델을 자동적으로 찾아주는 기법
3. 훈련 데이터의 잡음 줄임 (Remove errors / outliers)

How to solve Underfitting
1. 파라미터가 더 많은 복잡한 모델을 선택
2. 모델 제약을 줄이기 (regularization hyperparameter 줄이기)
3. 조기 종료 시점까지 충분히 학습 (overfitting 되기 전의 시점)
