# If

```
x = 5
y = x if x == 5 else 0
print(y)

x = 10
y = x if x == 5 else 0
print(y)


a = 75

if a > 50:
  if a < 100:
    print('50보다 크고 100보다 작군요')
  else:
    print('100보다 크군요')
else:
  print('50보다 작군요')



score = int(input("점수를 입력하세요 : "))

if score >= 90:
  print('A')
elif score >= 80:
  print('B')
elif score >= 70:
  print('C')
elif score >= 60:
  print('D')
else:
  print('F')

print(score)
```

# While

```
i = 0
while i < 100:
  print(i, "Hello, World!", end = " ")
  i+= 1


import random
i = 0
while i != 3:
  i = random.randint(0, 9)
  print(i)


count = 3
i = 1
while 1:
  print(i)
  if i == count:
    break;
  i += 1
```

# Break, Continue

Break: loop 완전 탈출

```
for i in range(1, 100):
  print("for 문을 %d 번 실행했습니다." %i)
  break


```

Continue: loop 한번만 탈출

```
hap, i = 0,0

for i in range(1, 101):
  if i % 3 == 0:
    continue

  hap +=i

print("1~100의 합계(3의 배수 제외): %d" %hap)
```
